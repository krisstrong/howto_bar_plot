# From Raw to Refined – A Step-by-Step Bar Plot Transformation in Python
This notebook walks through the full process of improving a basic bar plot using `matplotlib` — starting from a default plot, stripping it down to only the bars, and incrementally refining it using better colour, layout, and labels.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create dummy sales data
df = pd.DataFrame({
    'Product': ['Phones', 'Cars', 'Forks', 'Trucks', 'Mugs'],
    'Sales': [100, 180, 130, 90, 110]
})

df.head()
```

## Step 1: Default Plot

This is the plot you get with minimal code. It’s functional but unpolished and your Director wll be unimpressed. We can do better.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.bar(df['Product'], df['Sales'], color='Blue')
plt.title('Monthly Sales')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.show()
```

## Step 2: Simplify the Plot

In this step, we will remove all non-essential elements, leaving only the bars. This minimalist approach ensures that the focus remains solely on the data, highlighting that one bar is higher even without additional context.

```python
fig, ax = plt.subplots(figsize=(8, 5))

ax.bar(df['Product'], df['Sales'], color='black')

# Remove all visual clutter
ax.set_xticks([])
ax.set_yticks([])
for spine in ax.spines.values():
    spine.set_visible(False)

ax.set_title('')
ax.set_xlabel('')
ax.set_ylabel('')

plt.tight_layout()
plt.show()
```

## Step 3: Add Style and Meaning

We now start adding back structure with purpose:

- Custom colour
- Axis ticks and labels
- A clear title and axis labels

```python
import matplotlib.pyplot as plt
import pandas as pd

fig, ax = plt.subplots(figsize=(8, 5))

# Styled bar chart
ax.bar(df['Product'], df['Sales'], color='#1f77b4', linewidth=2.5)

# Show ticks
ax.set_xticks(df['Product'])
ax.set_xticklabels(df['Product'], fontsize=10)
#ax.set_yticks([100, 150, 200])
#ax.set_yticklabels(['100', '150', '200'], fontsize=10)

# Add labels
ax.set_title('Product Sales in 2023', fontsize=14, fontweight='bold', pad=15)
ax.set_xlabel('Product', fontsize=12)
ax.set_ylabel('Sales', fontsize=12)

# Style axes
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.grid(axis='y', linestyle='--', alpha=0.3)

plt.tight_layout()
plt.show()
```

The visualization has improved significantly, but there are still opportunities for further refinement.

## Step 4: Final Polish

We finalize with:

- Switch y and x axis - Generally, we read left to right
- Add data values to the end of the bars and remove the values in the x axis - this cleans up our plot
- Reorder values to improve presentation
- A subtitle or source note
- Remove the grid
- Fine-tuned layout

```python
import matplotlib.pyplot as plt
import pandas as pd


# Reorder the values
df = df.sort_values('Sales', ascending=True)

fig, ax = plt.subplots(figsize=(8, 5))

# Bar Plot
ax.barh(df['Product'], df['Sales'], color='#1f77b4', linewidth=0)

# Add data values at the inside end of the bars
for index, value in enumerate(df['Sales']):
    ax.text(value - 7.5, index, f'{value}', color='White', va='center', fontsize=10)

# Customize ticks
ax.set_yticks(range(len(df['Product'])))
ax.set_yticklabels(df['Product'], fontsize=10)  

ax.tick_params(axis='x', which='both', length=0, labelbottom=False)
ax.tick_params(axis='y', which='both', length=0)

# Labels and title
ax.set_title('Product Sales – 2023', fontsize=16, fontweight='bold', pad=15)
ax.set_xlabel('Sales (CAD)', fontsize=12)
ax.set_ylabel('', fontsize=12)

# Grid & background
ax.grid(axis='x', linestyle='--', alpha=0.0)
ax.set_facecolor('#fafafa')

# Remove clutter
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

# Optional: Add a source note
plt.figtext(0.9, 0.01, 'Source: Dummy data', ha='right', fontsize=8, color='gray')

plt.tight_layout()
plt.show()
```

To enhance the visual focus of our bar plot, we can switch the plot's theme to a dark mode version.

```python
import matplotlib.pyplot as plt
import pandas as pd


# Reorder the values
df = df.sort_values('Sales', ascending=True)

fig, ax = plt.subplots(figsize=(8, 5))

# Bar Plot
ax.barh(df['Product'], df['Sales'], color='#17becf', linewidth=0)

# Add data values at the inside end of the bars
for index, value in enumerate(df['Sales']):
    ax.text(value - 7.5, index, f'{value}', color='#1E1E2E', va='center', fontsize=10)

# Customize ticks
ax.set_yticks(range(len(df['Product'])))
ax.set_yticklabels(df['Product'], fontsize=10, color='White')  

ax.tick_params(axis='x', which='both', length=0, labelbottom=False)
ax.tick_params(axis='y', which='both', length=0)

# Labels and title
ax.set_title('Product Sales – 2023', fontsize=16, color = 'white', fontweight='bold', pad=15)
ax.set_xlabel('Sales (CAD)', fontsize=12, color = 'white',)
ax.set_ylabel('', fontsize=12)

# Grid & background
ax.grid(axis='x', linestyle='--', alpha=0.0)

fig.patch.set_facecolor('#1E1E2E')
ax.set_facecolor('#1E1E2E')

# Remove clutter
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

# Optional: Add a source note
plt.figtext(0.9, 0.01, 'Source: Dummy data', ha='right', fontsize=8, color='gray')

plt.tight_layout()
plt.show()
```

Notice how the focus remains on the narrative and the direction it provides. The narrative now stands out clearly, ensuring that your Director will immediately grasp the plot and its significance.
